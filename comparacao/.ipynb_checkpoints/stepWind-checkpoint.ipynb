```python
import pandas as pd
import matplotlib.pyplot as plt

dataCurrent = {'PQ':pd.read_csv("csvs/PQ_current.csv"),
               'Te':pd.read_csv("csvs/Te_current.csv"),
               'Wrm':pd.read_csv("csvs/Wrm_current.csv"),
               'I':pd.read_csv("csvs/I_current.csv"),
               'V':pd.read_csv("csvs/V_current.csv"),
               'Vcc':pd.read_csv("csvs/Vcc_current.csv")}

dataVoltage = {'PQ':pd.read_csv("csvs/PQ_voltage.csv"),
               'Te':pd.read_csv("csvs/Te_voltage.csv"),
               'Wrm':pd.read_csv("csvs/Wrm_voltage.csv"),
               'I':pd.read_csv("csvs/I_voltage.csv"),
               'V':pd.read_csv("csvs/V_voltage.csv"),
               'Vcc':pd.read_csv("csvs/Vcc_voltage.csv")}
```

```python
# Declaração da Figura:
fig, [ax1, ax2] = plt.subplots(2, 1, sharex=True)
fig.set_size_inches(8, 6)

# Plots:
ax1.plot(dataVoltage["PQ"]["time"], dataVoltage["PQ"]["voltageSource.S.re"],
         label="Voltage control", color='r')
ax1.plot(dataCurrent["PQ"]["time"], dataCurrent["PQ"]["voltageSource.S.re"], 
         label="Current control", color='b', linestyle='-.')

ax1.set(xlabel='',    ylabel='$P_g$ $(pu)$')

ax2.plot(dataVoltage["PQ"]["time"], dataVoltage["PQ"]["voltageSource.S.im"], 
         label="Voltage control", color='r')
ax2.plot(dataCurrent["PQ"]["time"], dataCurrent["PQ"]["voltageSource.S.im"], 
         label="Current control", color='b', linestyle='-.')

ax2.set(xlabel='tempo (s)',    ylabel='$Q_g$ $(pu)$')

ax1.grid(True, linestyle='dotted')
ax2.grid(True, linestyle='dotted')

ax1.legend(loc='upper right')
plt.show()

# fig.savefig("/content/drive/MyDrive/IC/grafics/clagtee2/Psr.pdf", bbox_inches='tight', pad_inches=0)
```
