```python
import pandas as pd
import matplotlib.pyplot as plt

data1 = pd.read_csv("csvs/PQ_current.csv")
data2 = pd.read_csv("csvs/Te_current.csv")
data3 = pd.read_csv("csvs/Wrm_current.csv")
data4 = pd.read_csv("csvs/I_current.csv")

data5 = pd.read_csv("csvs/PQ_voltage.csv")
data6 = pd.read_csv("csvs/Te_voltage.csv")
data7 = pd.read_csv("csvs/Wrm_voltage.csv")
data8 = pd.read_csv("csvs/I_voltage.csv")
```

```python
# Declaração da Figura:
fig, [ax1, ax2] = plt.subplots(2, 1, sharex=True)
fig.set_size_inches(8, 6)

# Plots:
ax1.plot(data5["time"], data5["voltageSource.S.re"], label="Voltage control", color='r')
ax1.plot(data2["time"], data1["voltageSource.S.re"], label="Current control", color='b', linestyle='-.')
ax1.set(xlabel='',    ylabel='$P_g$ $(pu)$')

ax2.plot(data5["time"], data5["voltageSource.S.im"], label="Voltage control", color='r')
ax2.plot(data1["time"], data1["voltageSource.S.im"], label="Current control", color='b', linestyle='-.')

ax2.set(xlabel='tempo (s)',    ylabel='$Q_g$ $(pu)$')

ax1.grid(True, linestyle='dotted')
ax2.grid(True, linestyle='dotted')

ax1.legend(loc='upper right')
plt.show()

# fig.savefig("/content/drive/MyDrive/IC/grafics/clagtee2/Psr.pdf", bbox_inches='tight', pad_inches=0)
```
